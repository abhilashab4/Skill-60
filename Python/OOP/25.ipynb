```python
#Multiple Inherirance, Abstract Method, 



class A:
    def greet(self):
        print("Hello from A")

class B(A):
    def greet(self):
        print("Hello from B")

class C(A):
    def greet(self):
        print("Hello from C")

class D(B,C):
    pass

def main():
    d=D()
    d.greet()
    
if __name__=="__main__":
    main()
```

```python
from abc import ABC,abstractmethod
class Shape(ABC):
    @abstractmethod
    def area(self):
        pass
    
    @abstractmethod
    def perimeter(self):
        pass

class Rectangle(Shape):
    def __init__(self,width,height):
        self.width=width
        self.height=height
    
    def area(self):
        return self.width*self.height
    
    def perimeter(self):
        return 2*(self.width+self.height)

class Circle(Shape):
    def __init__(self,radius):
        self.radius=radius
    
    def area(self):
        return 3.1416*self.radius**2
    
    def perimeter(self):
        return 2*3.1416*self.radius

def main():
    shapes=[Rectangle(3,4),Circle(5)]
    for shape in shapes:
        print(f"{shape.__class__.__name__}: Area = {shape.area()}, Perimeter = {shape.perimeter()}")

if __name__ == "__main__":
    main()
```

```python
# Instructions
# 1. Abstract Base Class: Animal (Abstraction)
# Create an abstract base class Animal with:
# Protected Attributes:
# _name: The animal’s name (string).
# _health: Health level (0–100, float).
# _hunger: Hunger level (0–100, float).
# _habitat_type: The type of habitat required (string, e.g., "Savanna").

 
# Abstract Methods:
# make_sound(self): Defines the animal’s unique sound.

 
# feed(self, food_amount): Defines how the animal eats and updates hunger/health.

 
# Concrete Methods:
# get_health(), get_hunger(), 
 
# get_habitat_type(): Getters for protected attributes.

 
# decrease_health(self, amount): Reduces _health (with a minimum of 0).

 
# increase_hunger(self, amount): Increases _hunger (with a maximum of 100).


# 2. Animal Subclasses (Inheritance and Polymorphism)

# Create at least three subclasses of Animal:

# Lion:

# make_sound: Prints "Roar!".

 
# feed: Reduces _hunger by food_amount * 2 (lions eat a lot) and increases _health by food_amount * 0.5.

 
 
# Elephant:

# make_sound: Prints "Trumpet!".

 
# feed: Reduces _hunger by food_amount * 1.5 and increases _health by food_amount * 0.3.

 
 
# Parrot:

# make_sound: Prints "Squawk!".

 
# feed: Reduces _hunger by food_amount * 0.5 (parrots eat less) and increases _health by food_amount * 0.2.


# 3. Abstract Base Class: Habitat (Abstraction)
# Create an abstract base class Habitat with:
# Protected Attributes:
# _type: Habitat type (string, e.g., "Savanna").

 
# _condition: Condition level (0–100, float, representing cleanliness/maintenance).

 
# _animals: List of animals in the habitat.

 
 
# Abstract Method:
# affect_animal(self, animal): Defines how the habitat impacts an animal’s health.

 
 
# Concrete Methods:
# add_animal(self, animal): Adds an animal if its _habitat_type matches.

 
# get_condition(): Getter for _condition.

 
# degrade(self, amount): Reduces _condition (e.g., due to animal activity).
```

```python
from abc import ABC, abstractmethod

class Animal(ABC):
    def __init__(self, name: str, health: float, hunger: float, habitat_type: str):
        self._name = name
        self._health = max(0, min(health, 100))  # Ensuring health is within 0-100
        self._hunger = max(0, min(hunger, 100))  # Ensuring hunger is within 0-100
        self._habitat_type = habitat_type
    
    @abstractmethod
    def make_sound(self):
        """Defines the animal’s unique sound."""
        pass
    
    @abstractmethod
    def feed(self, food_amount: float):
        """Defines how the animal eats and updates hunger/health."""
        pass
    
    def get_health(self):
        return self._health
    
    def get_hunger(self):
        return self._hunger
    
    def get_habitat_type(self):
        return self._habitat_type
    
    def decrease_health(self, amount: float):
        self._health = max(0, self._health - amount)
    
    def increase_hunger(self, amount: float):
        self._hunger = min(100, self._hunger + amount)



class Lion(Animal):
    def make_sound():
        print("Roar!")

    def feed(self, food_amount: float):
        self._hunger = max(0, self._hunger - food_amount * 2)
        self._health = min(100, self._health + food_amount * 0.5)


class Elephant(Animal):
    def make_sound(self):
        print("Trumpet!")

    def feed(self, food_amount: float):
        self._hunger = max(0, self._hunger - food_amount * 1.5)
        self._health = min(100, self._health + food_amount * 0.3)


class Parrot(Animal):
    def make_sound():
        print("Squawk!")

    def feed(self, food_amount: float):
        self._hunger = max(0, self._hunger - food_amount * 0.5)
        self._health = min(100, self._health + food_amount * 0.2)


class Habitat(ABC):

    def __init__(self, habitat_type: str, condition: float):
        self._type = habitat_type
        self._condition = max(0, min(condition, 100))  # Ensuring condition is within 0-100
        self._animals = []
    
    @abstractmethod
    def affect_animal(self, animal: Animal):
        """Defines how the habitat impacts an animal’s health."""
        pass
    
    def add_animal(self, animal: Animal):
        if animal.get_habitat_type() == self._type:
            self._animals.append(animal)
    
    def get_condition(self):
        return self._condition
    
    def degrade(self, amount: float):
        self._condition = max(0, self._condition - amount)

    
def main():

    lion = Lion("Leo", 80, 30, "Savanna")
    elephant = Elephant("Dumbo", 90, 40, "Savanna")
    parrot = Parrot("Polly", 70, 20, "Jungle")

    savanna = Habitat("Savanna", 80)
    jungle = Habitat("Jungle", 70)

    savanna.add_animal(lion)
    savanna.add_animal(elephant)
    jungle.add_animal(parrot)

    print(savanna.get_condition())
    savanna.degrade(10)
    print(savanna.get_condition())

    for animal in savanna._animals:
        savanna.affect_animal(animal)

    for animal in jungle._animals:
        jungle.affect_animal(animal) 


if __name__ == "__main__":
    main()
```
